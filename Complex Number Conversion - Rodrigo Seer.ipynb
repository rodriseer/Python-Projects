```python
#Rodrigo Seer


#intro to program
programmer = str("Rodrigo Seer")
user_name = input("Hello! May I have your name, please?\nName:")
print("Hi", user_name,"! I am a number conversion program!\nTo start, select one of the options below:\n")


#function creation "conversion", it selects the options for conversion
def conversion():
#while if operations for selections
  while True:
    print("1. Decimal to hexadecimal")
    print("2. Decimal to octal")
    print("3. Octal to decimal")
    print("4. Binary to decimal")
    print("5. Exit")
#choice data
    choice = input("Enter your choice: ")
    try:
      choice = int(choice)
#making sure the user doesnt input text in this part
    except ValueError:
      print("Please enter a valid option\n")
      continue
    if choice <= 0 or choice > 5:
      print("Please, select a valid option, try again.\n")
      continue
    return choice


#function creation "get number", to input number, with test to make sure the user doesnt input more than 2048 or less than 0, save the number in integer, loop if user innputs out of range number
def get_number(choice):
  while True:
    number = input("Enter the number for conversion: ")
    try:
#saves as integer
      number = int(number)
#tests if numbers out of range for octal, binary and out of range
      if 0 <= number <= 2048:
        if choice == 3 and not is_octal(number):
          print("Not a valid octal number!")
          continue
        if choice == 4 and not is_binary(number):
          print("Not a valid binary number!")
          continue
        return number
#loops again if user input wrong value
      else:
        print("Number should be between 0 and 2048, select option again.\n")
#test if user inputs letters or phrases
    except ValueError:
      print("Please enter a valid number, not a text string.")


#function creation to test if number is octal or not, that is, in range of 0-7
def is_octal(num):
#convert number to string
  for number_sequence in str(num):
    if number_sequence not in '01234567':
      return False
  return True


#function creation to test if number is binary
def is_binary(num):
#convert number to string
  for number_sequence in str(num):
    if number_sequence not in '01':
      return False
  return True


#function creation for final conversions
def operation(choice, number):
#decimal to hexa
  if choice == 1:
    print("The decimal", number, "is ", hex(number), "in hexadecimal\nWould you like to do another conversion? Select '5' to exit program.\n")
#decimal to octal
  elif choice == 2:
    print("The decimal", number, "is ", oct(number), "in octal\nWould you like to do another conversion? Select '5' to exit program.\n")
#octal to decimal
  elif choice == 3:
    if is_octal(number):
      convert_octal = int(str(number), 8)
      print ("The Octal", number, "is", convert_octal, "in decimal\nWould you like to do another conversion? Select '5' to exit program.\n")
    else:
      print("Not a valid octal number!")
#binary to decimal
  elif choice == 4:
    if is_binary(number):
      convert_binary = int(str(number), 2)
      print("The binary,", number, "is", convert_binary, "in decimal\nWould you like to do another conversion? Select '5' to exit program.\n")
  else:
    print("Not a valid binary number!")


#calling functions and allowing the user to choose another conversion option if desired
while True:
  choice = conversion()
  if choice == 5:
    print("Goodbye", user_name, "!\nProgram made by", programmer)
    break
  number = get_number(choice)
  operation(choice, number)





```
